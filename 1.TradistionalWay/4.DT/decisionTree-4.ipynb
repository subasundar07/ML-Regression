```python
import pandas as pd
detaset=pd.read_csv("50_Startups.csv")
```

```python
detaset
```

```python
detaset=pd.get_dummies(detaset, dtype=int, drop_first=True)
```

```python
detaset
```

```python
detaset.columns
```

```python
independent=detaset[['R&D Spend', 'Administration', 'Marketing Spend', 
       'State_Florida', 'State_New York']]
```

```python
independent
```

```python
dependent=detaset[['Profit']]
```

```python
dependent
```

```python
#split train and test deta
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#from sklearn.tree import DecisionTreeRegressor
#regressor=DecisionTreeRegressor(criterion='friedman_mse', splitter='random')
#regressor=regressor.fit(X_train,Y_train)
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor=regressor.fit(X_train,Y_train)
```

```python
#import matplotlib_pyplot as plt
#from sklearn import tree 
#tree.plot_tree()
#plt.show()
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
#Evaluate the model
Y_pred=regressor.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(Y_pred,Y_test)

```

```python
r_score
```

```python
#save Model
import pickle
filename="finale_model.sav"
pickle.dump(regressor,open(filename,'wb'))
loaded_model=pickle.load(open(filename,'rb'))
```
