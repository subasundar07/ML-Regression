```python
#data collection
import pandas as pd #--->pandas is a library it maily used for table deta
dataset=pd.read_csv("Salary_data.csv")#read is used to reading the dataset
```

```python
dataset
```

```python
#split intput & output
indepentent=dataset[["YearsExperience"]] #-->creating A input variable to find the inputdeta
indepentent
```

```python
dependent =dataset[["Salary"]] #--->creating a variable to find the output deta
dependent
```

```python
#split test and train deta
from sklearn.model_selection import train_test_split #--->using the sklearn library ,for split test and train deta using a model_selection library
#and import train_test_split library
X_train , X_test , Y_train , Y_test = train_test_split (indepentent,dependent,test_size=0.30,random_state=0) #i|p & o|p deta test,train deta =input,ouput,test deta ,random state
```

```python
#Model Creation
from sklearn.linear_model import LinearRegression #choose a sklearn.linearmodel library and import a lineaar regression library for creating a model
regressor=LinearRegression() #-->crating a variable for assign the linear regression model
regressor.fit(X_train , Y_train) #-->using fit keyword to to acess the x and y train deta
```

```python
weight=regressor.coef_ #--->for weigth using coef_keyword
weight
```

```python
Bias=regressor.intercept_#--->for bias using intercept_keyword
Bias
```

```python
#Evaluation metrix
Y_Pred=regressor.predict(X_test) 

```

```python
from sklearn.metrics import r2_score #-->to evaluation metrics using keyword metircs and and import library r2_score
r_score=r2_score(Y_test , Y_Pred)
```

```python
r_score
```

```python
pickle.dump(regressor,open(filename,'wb')) #-->wb is used to write the binary code
#pickle.dump(regressor,open(filename,'wb'))
```
