```python
import pandas as pd
detaset=pd.read_csv("insurance_pre.csv")
```

```python
detaset
```

```python
detaset=pd.get_dummies(detaset,dtype=int,drop_first=True)
```

```python
detaset
```

```python
detaset.columns
```

```python
independent=detaset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
independent
```

```python
dependent=detaset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='squared_error', max_features='sqrt', splitter='Random')
regressor=regressor.fit(X_train,Y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree 
tree.plot_tree(regressor)
plt.show()
```

```python
#evaluate the model
Y_pred=regressor.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(Y_pred,Y_test)
r_score
```

```python
#save the model
import pickle
filename="final_model.sav"
pickle.dump(regressor,open(filename,'wb'))


```
