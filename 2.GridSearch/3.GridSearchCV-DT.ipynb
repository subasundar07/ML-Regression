```python
import numpy as np
import matplotlib.pyplot as pd
import pandas as pd
```

```python
detaset=pd.read_csv('insurance_pre.csv')
detaset
```

```python
detaset=pd.get_dummies(detaset, dtype=int, drop_first=True)
detaset
```

```python
detaset.columns
```

```python
independent=detaset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]

```

```python
independent
```

```python
dependent=detaset[['charges']]
```

```python
dependent
```

```python
#split train and test deta
from sklearn.model_selection import train_test_split
X_train, X_test ,Y_train, Y_test= train_test_split(independent, dependent,test_size=1/3, random_state=0)
```

```python
#preprocessing
from sklearn.preprocessing import StandardScaler  
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)

```

```python
#model creation




from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
param_grid = {'criterion':['mse','mae','friedman_mse'],
'max_features': ['auto','sqrt','log2'],
'splitter':['best','random']}


```

```python

grid = GridSearchCV(DecisionTreeRegressor(), param_grid, refit = True, verbose = 3,n_jobs=-1)
```

```python
# fitting the model for grid search
grid.fit(X_train, Y_train)
```

```python
#choosing best model
#re=grid.cv_results_
#print=("best_model{}".format(grid.best_params_))

re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```
